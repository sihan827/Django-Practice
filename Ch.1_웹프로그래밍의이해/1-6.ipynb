## Chapter 01: 웹 프로그래밍의 이해

### 1.6 웹 어플리케이션 서버
- 서버를 다음과 같이 웹 서버와 웹 어플리케이션 서버로 세분화 가능

|구분|역할|프로그램명|
|----|----|---------|
|웹 서버|웹 클라이언트의 요청을 받아 처리하고 그 결과를 웹 클라이언트에 응답, 주로 정적인 HTML, CSS, Javascript 파일을 제공|Apache httpd, NGINX, lighttpd, IIS 등|
|웹 어플리케이션 서버|웹 서버로부터 동적 페이지 요청을 받아 처리하고 그 결과를 웹 서버로 반환, 동적 페이지 생성을 위한 프로그램 실행과 데이터베이스 연동 기능 처리|Apache Tomcat, JBoss, WebLogic, WebSphere, Jetty 등|

- 기술에 따른 웹 서버 기술의 변화는 다음과 같음

|웹 서버 등장|CGI 프로그램 등장|CGI 대안 기술 2가지|웹 어플리케이션 서버 등장|
|------------|----------------|-----------------|-----------------------|
|웹 서버|웹 서버 <-> CGI 프로그램|웹 서버 <-> 데몬 프로그램</br>웹 서버 내의 스크립트 엔진|웹 서버 <-> 웹 어플리케이션 서버|

#### 1.6.1 정적 페이지 vs 동적 페이지
- 정적 페이지(Static page)는 누가, 언제 요구하더라도 항상 같은 내용을 표시하는 웹 페이지를 의미
  - 사전에 HTML, CSS, Javascript 파일을 준비하여 동일한 리소스의 요청에 대해 항상 동일한 페이지 반환
- 동적 페이지(Dynamic page)는 동일한 리소스의 요청이더라도 누가, 언제, 어떻게 요구했냐에 따라 다른 내용이 표시되는 웹 페이지를 의미
  - 장바구니, 현재 시각을 보여주는 페이지 등
- 데이터베이스의 사용과 동적 페이지에 대한 수요가 늘어남에 따라 웹 서버와는 별도의 프로그램이 필요
- 따라서 CGI(Common Gateway Interface) 규격으로 웹 서버와 별도 프로그램 간의 정보 교환 규칙을 정의

#### 1.6.2 CGI 방식의 단점
- 전통적인 CGI 방식은 웹 서버가 C, C++, Perl 등으로 만들어진 CGI 프로그램을 호출하여 개별 프로세스를 생성
- 각 클라이언트의 요청에 독립 프로세스를 각각 생성하므로 처리 비용이 많아진다는 단점 존재

#### 1.6.3 CGI 방식의 대안 기술
1. 별도의 어플리케이션을 스크립트 언어로 작성하고 스크립트 엔진을 웹 서버에 내장시켜 개별 프로세스에 대한 부하를 줄임
  - 아파치 웹 서버의 mod_per, mod_php 등
  - 파이썬의 mod_wsgi 모듈
2. 어플리케이션 처리용 프로세스를 미리 데몬으로 가동시킨 후 웹 서버의 요청을 데몬에서 처리하여 프로세스 생성 부하를 줄임
  - 파이썬의 mod_wsgi 모듈
3. 데몬 처리 방식이 발전함에 따라 어플리케이션 전용 데몬인 어플리케이션 서버 방식이 개발됨
  - JSP(Java server page), ASP(Active server page) 등
  - 파이썬의 mod_wsgi, uwsgi, gunicorn 등

#### 1.6.4 어플리케이션 서버 방식
- 어플리케이션 서버 방식에서는 웹 서버가 직접 프로그램을 호출하지 않고 웹 어플리케이션 서버에 처리를 위임하여 간접적으로 어플리케이션 프로그램을 실행하고 그 결과를 전달받음
- 웹 서버는 정적 페이지를 처리하도록 하고 웹 어플리케이션 서버는 동적 페이지를 처리하도록 역할 분담하는 것이 중요
- 웹 서버의 기능은 다음과 같음
  - 정적 페이지 및 캐시, 프록시 등 추가 기능 제공
  - 동시 접속 허용 클라이언트수 제한
  - 처리 프로세스 관리
  - 요청/응답에 대한 로그 기록
  - 인증 제어 및 암호화 처리 등
- 웹 어플리케이션 서버 또한 웹 서버의 기능을 제공하지만 성능과 안정성이 떨어지므로 대규모 운용 환경에서는 웹 서버의 기능을 잘 사용하지 않음

#### 1.6.5 웹 서버와의 역할 구분
- 웹 서버와 웹 어플리케이션 서버를 하나의 하드웨어 박스에 탑재할 수 있으며 이 경우 서비스 운용 관리 측면에서 더 편리함
- 각 서버의 하드웨어 박스를 나누어 운용할 경우 메모리 효율을 더 높일 수 있으며 이를 위해서는 정적 페이지와 동적 페이지의 요청 건수를 분석해야 함
